```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity  # Added the 's' to metrics
```

```python
# 1. Our Internal Knowledge Base (The Corpus)
# This mimics the "Vector Document Store" mentioned in your assignment guidelines
knowledge_base = {
    "What is the grading policy?": "The grading policy requires a minimum of 50% to pass the internship tasks.",
    "When is the final assignment due date?": "The final submission deadline for all advanced tasks is June 30th, 2026.",
    "How do I submit my completed projects?": "You must upload your code to GitHub and submit the repository link via Google Classroom.",
    "What technologies are we using?": "We are using Python, Scikit-learn, PyTorch, and Hugging Face Transformers."
}
# Extract just the questions to use as our searchable document texts
documents = list(knowledge_base.keys())
```

```python
# 2. Setup a lightweight Text Vectorizer
vectorizer = TfidfVectorizer()
document_vectors = vectorizer.fit_transform(documents)
```

```python
# 3. Chatbot History Memory System
# A simple list to keep track of what the user asks during the session
conversation_history = []

def context_aware_chatbot(user_question):
    # Store question in conversational history memory
    conversation_history.append(f"User: {user_question}")
    
    # Convert the user's question into math features
    query_vector = vectorizer.transform([user_question])
    
    # Calculate similarity scores between the question and our knowledge base docs
    similarity_scores = cosine_similarity(query_vector, document_vectors).flatten()
    
    # Find the document with the highest matching score
    best_match_idx = np.argmax(similarity_scores)
    highest_score = similarity_scores[best_match_idx]
    
    # Threshold check: If the match score is too low, the bot doesn't know the answer
    if highest_score > 0.20:
        matched_question = documents[best_match_idx]
        bot_response = knowledge_base[matched_question]
    else:
        bot_response = "I'm sorry, I couldn't find that specific information in my knowledge base."
        
    # Store bot response in history memory
    conversation_history.append(f"Bot: {bot_response}")
    return bot_response
```

```python
# --- Test the Chatbot Live ---
print("Context-Aware Chatbot Initialized!\n" + "="*40)

# Simulate a conversation
questions = [
    "Hey, when is the final assignment due date?",
    "Okay, and how do I submit my completed projects?",
    "Can you order a pizza for me?"
]

for q in questions:
    print(f"\nUser asks: '{q}'")
    reply = context_aware_chatbot(q)
    print(f"Chatbot Reply: {reply}")
```

```python
# 4. Display the complete conversational history (Memory check)
print("\n" + "="*20 + " CONVERSATION HISTORY MEMORY " + "="*20)
for line in conversation_history:
    print(line)
```
