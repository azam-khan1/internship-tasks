### Importing Visualizing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Iris.csv')       #reading csv file 
print(df.columns)          #displaying columns names               

```

```python
print(df.shape)             #displaying shape 
df.info()
```

```python
df.head()                  #displaying first 5 records
```

```python
df.describe()               #displaying statistical information
```

```python
x = df['SepalLengthCm']
y = df['SepalWidthCm']
plt.scatter(x,y)
plt.show()
```

```python


# 1. Define the first subplot (1 row, 2 columns, 1st plot)
plt.subplot(1, 2, 1)
plt.hist(x, color='blue')
plt.title('Sepal Length')
plt.subplot(1, 2, 2)
plt.hist(y, color='orange')
plt.title('Sepal Width')
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(df)
plt.show()
```
